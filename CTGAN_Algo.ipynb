```python
combine_df = load_demo()

combine_df = combine_df.replace([np.inf, -np.inf], 0)

#label encoder is used to convert categorical date to numeric data
#here we have converted last label : marker (natural,attack) to (0,1)

# Creating a instance of label Encoder.
le = LabelEncoder()
 
# Using .fit_transform function to fit label
# encoder and return encoded label
label = le.fit_transform(combine_df['marker'])
 
# printing label
label

# removing the column 'marker' from df
# as it is of no use now.
combine_df.drop("marker", axis=1, inplace=True)

# Appending the array to our dataFrame
# with column name 'label'
combine_df["marker"] = label

# printing Dataframe
combine_df

#storing as a CSV file
combine_df.to_csv('preData15.csv', index=False)

#preData files are used for testing. Its a preprocessed file.
```
